```python
import pandas as pd
import numpy as np

def lastValue(x):
    #looks for last non null value and returns NaN if empty
    try:
        return x.dropna().iloc[-1]
    except:
        return x.iloc[-1]
```

```python
miles =miles.sort_values(["period_id","mile_id"])
```

```python
g = miles.groupby("period_id" ).agg({"mileage": lambda x: x.iloc[-1], "collision_mileage": lambda x: x.iloc[-1]})
```

```python
g2 = miles.groupby("period_id").agg({"mileage": lambda x: x.diff().sum(), "collision_mileage": lambda x: x.diff().sum()})
```
