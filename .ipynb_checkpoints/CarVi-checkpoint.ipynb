# TABLE JOINS

```python
# check relationship between period, motions, miles 
print len(pd.merge(period, motions, how='inner', on='period_id'))
print len(pd.merge(period, miles, how='inner', on='period_id'))
print len(pd.merge(motions, miles, how='inner', on='motion_id'))
```

```python
# check relationship between motions and colision, departure, front, suddens
print len(pd.merge(motions, collision, how='inner', on='motion_id'))
print len(pd.merge(motions, departure, how='inner', on='motion_id'))
print len(pd.merge(motions, front, how='inner', on='motion_id'))
print len(pd.merge(motions, suddens, how='inner', on='motion_id'))
```

```python
# check relationship between miles and colision, departure, front, suddens
print len(pd.merge(miles, collision, how='inner', on='motion_id'))
print len(pd.merge(miles, departure, how='inner', on='motion_id'))
print len(pd.merge(miles, front, how='inner', on='motion_id'))
print len(pd.merge(miles, suddens, how='inner', on='motion_id'))
```

```python
# check that collision, front, departure, and suddents don't have overlapping motion_id
print len(pd.merge(collision, front, how='inner', on='motion_id'))
print len(pd.merge(collision, departure, how='inner', on='motion_id'))
print len(pd.merge(collision, suddens, how='inner', on='motion_id'))
print len(pd.merge(front, departure, how='inner', on='motion_id'))
print len(pd.merge(front, suddens, how='inner', on='motion_id'))
print len(pd.merge(departure, suddens, how='inner', on='motion_id'))
```
